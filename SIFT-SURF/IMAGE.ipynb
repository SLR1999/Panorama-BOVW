```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./Bikes/0007.jpg")

sift = cv2.xfeatures2d.SIFT_create()

keyPoints, descriptor = sift.detectAndCompute(image, None)

key_points_image = cv2.drawKeypoints(image, keyPoints, outImage=None)
plt.imshow(key_points_image, aspect='auto')
```

```python
surf = cv2.xfeatures2d.SURF_create()

keyPoints_surf, descriptor_surf = surf.detectAndCompute(image, None)

key_points_image = cv2.drawKeypoints(image, keyPoints_surf, outImage=None)
plt.imshow(key_points_image, aspect='auto')
```
